```python
# convert rating value from type str to int
df['Rating_Int'] = df['Rating (Num)'].apply(lambda x: int(x))
df
```

```python
df.describe()
```

```python
def sentiment_category(rating):
    if rating >= 9:
        return "Promoter"
    elif rating >= 7:
        return "Passive"
    else:
        return "Detractor"

# sentiment category column
df['Sentiment'] = df["Rating_Int"].apply(sentiment_category)
df
```
